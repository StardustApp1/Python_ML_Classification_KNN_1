```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Fetching data
data = pd.read_csv(r'E:\4. Python\Datasets & GitHub Projects\2. Classification Algorithms\KNN\Classified Data', index_col=0)
```

```python
data.head(2)
```

# Standardizing the Variables

```python
# Step 1 - importing StandardScaler library from sklearn.preprocessing
from sklearn.preprocessing import StandardScaler
```

```python
# Step 2 - assigning variable 'scaler' for StandardScaler
scaler = StandardScaler()
```

```python
# Step 3 -
scaler.fit(data.drop('TARGET CLASS', axis=1))
```

```python
# Step 4 - Transforming the data
scaled_features = scaler.transform(data.drop('TARGET CLASS', axis=1))
```

```python
scaled_features
```

```python
# Step 5 - Converting to Dataframe and assigning columns name
data_feat = pd.DataFrame(scaled_features, columns=data.columns[:-1])
data_feat.head()
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = data_feat
y = data['TARGET CLASS']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30) 
```

# Using KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

```python
predictions = knn.predict(X_test)
```

```python
predictions
```

## Predictions and Evaluations

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
print(classification_report(y_test, predictions))
```

## Choosing a K Value

```python
error_rate = []

# executing k value loop takes time
for i in range(1,60):

    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.plot(range(1,60), error_rate)
```

```python
# for K=1
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
pred1 = knn.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix

print('for K=1')
print(classification_report(y_test, pred1))
print('\n')
print(confusion_matrix(y_test, pred1))


```

```python
# for K=1
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=30)
knn.fit(X_train, y_train)
pred1 = knn.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix

print('for K=1')
print(classification_report(y_test, pred1))
print('\n')
print(confusion_matrix(y_test, pred1))
```
